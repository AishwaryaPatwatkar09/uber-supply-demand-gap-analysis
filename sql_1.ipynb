```python
df.columns
```

```python
query = """
SELECT [Request timestamp], COUNT(*) as request_count
FROM df
GROUP BY [Request timestamp]
"""
ps.sqldf(query, locals())
```

```python
query = """
SELECT [Time slot], COUNT(*) as cancel_count
FROM df
WHERE Status = 'Cancelled'
GROUP BY [Time slot]
ORDER BY cancel_count DESC
"""
ps.sqldf(query, locals())
```

```python
query = """
SELECT Status, COUNT(*) as count
FROM df
GROUP BY Status
ORDER BY count DESC
"""
ps.sqldf(query, locals())
```

```python
query = """
SELECT [Pickup point], Status, COUNT(*) as count
FROM df
GROUP BY [Pickup point], Status
ORDER BY [Pickup point], Status
"""
ps.sqldf(query, locals())
```

```python
result = ps.sqldf(query, locals())
result.to_csv("Query3_StatusWise.csv", index=False)
```
